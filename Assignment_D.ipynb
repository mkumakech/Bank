```python
#import libraries
import pandas as pd
import numpy as np
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
#read the data file
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
#check the data 
concrete_data.shape  #number of rows and columns 
```

```python
concrete_data.describe() #some values about the data
```

```python
concrete_data.isnull().sum()
```

```python
#split the data into target strength and predictors
concrete_data_columns = concrete_data.columns
```

```python
target = concrete_data['Strength']
target.head(10)
```

```python
predictors = concrete_data.iloc[:, :-1]
predictors.head(10)
```

```python
predictors_norm = (predictors - predictors.mean())/predictors.std()
predictors_norm.head(10)
```

```python
n_cols = predictors_norm.shape[1]
def regression_model4():
    #create model4
    #Three hidden layers
    model4 = Sequential()
    model4.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model4.add(Dense(10, activation='relu'))
    model4.add(Dense(10, activation='relu'))
    model4.add(Dense(1))
    model4.compile(optimizer='adam', loss='mean_squared_error')
    return model4
#build the model
model4 = regression_model4()
```

```python
mean_squared_error = [] #list of the mean squared errors
for i in range(50):
      #load data --> 0.3 for testing
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3)
    
    res = model4.fit(X_train, y_train, epochs=50, verbose=0, validation_data=(X_test, y_test))
   
    error = res.history['val_loss'][-1]
    
    mean_squared_error.append(error)
    print('Step {}: mean squared error--> {}'.format(i+1, error))
    
#The mean and the standard deviation of the mean squared errors in case D is less than in case A, B and C. 
```

```python
print('The mean of the mean squared errors: {}'.format(np.mean(mean_squared_error)))
print('The standard deviation of the mean squared errors: {}'.format(np.std(mean_squared_error)))
```
